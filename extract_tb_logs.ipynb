```python
import pandas as pd
from matplotlib import pyplot as plt
from glob import glob
from torch.utils.tensorboard import SummaryWriter
from tensorboard_reducer import load_tb_events, reduce_events, write_csv, write_tb_events
import numpy as np
import tensorflow as tf
def load_tensorboard(event_file):
    data = {}
    for e in tf.train.summary_iterator(event_file):
        for v in e.summary.value:
            if v.tag not in data:
                data[v.tag] = []
            data[v.tag].append((e.step, v.simple_value))
    return data
project="MTL-sup-2opt-10x_lr"
versions=[129]+list(range(134,144))
filters=['val_accuracy_lab1','loss_recon','loss_seg','sigma_recon']

for ver in versions:
    in_dirs = f"/mnt/Data/nathan/DeepSeg/tb_logs/my_model/version_{ver}"
    out_dir = f"/mnt/Data/nathan/DeepSeg/tb_logs_reduced/{project}/version_{ver}"
    overwrite = False
    events_dict = load_tb_events(in_dirs,handle_dup_steps='keep-first')
    for key in list(events_dict.keys()):
        if key not in filters:
            del events_dict[key]
    write_tb_events(events_dict,out_dir,overwrite=True)




    
```
