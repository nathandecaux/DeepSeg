```python
import pandas as pd
import json
import numpy as np
import sys
import os
import pytorch_lightning as pl
from pytorch_lightning.loggers import TensorBoardLogger

dir= os.getcwd()
sys.path.append(dir)
sys.path.append(dir.replace('/benchmark',''))
import gc
import torch
def wipe_memory():
    gc.collect()
    torch.cuda.empty_cache()
from models.MTL import MTL
from data.PlexDataModule import ACDCDataModule

from predict import save_preds
torch.cuda.empty_cache()

wipe_memory()
with open(f'{dir}/results/results.json','r') as f:
    df=json.load(f)

exp0=df['ACDC']['5']['UNet'][0]
dm=ACDCDataModule(subject_ids=exp0['subject_ids'],test_ids=exp0['test_ids'],batch_size=8)
neptune_logger = TensorBoardLogger("tb_logs", name="MTL",log_graph=True)
trainer=pl.Trainer(gpus=1,max_epochs=50,logger=neptune_logger)
model=MTL(n_classes=4,n_features=8,taa=True,learning_rate=1e-3)
trainer.fit(model,datamodule=dm)
preds=trainer.test(model=model,datamodule=dm)
print(preds)
```
